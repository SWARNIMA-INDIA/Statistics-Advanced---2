1. What is hypothesis testing in    statistics?
  - Hypothesis testing is a statistical technique used to make decisions about a population based on sample data. It helps us determine whether a given assumption or claim about a population parameter is supported by the sample data or not.
  - In hypothesis testing, we set up hypotheses and use statistical tests to decide whether to accept or reject the assumption at a certain level of significance.

2. What is the null hypothesis, and how does it differ from the alternative
hypothesis?
  - Null Hypothesis (H₀):
It is the default assumption that there is no effect, no difference, or no relationship in the population.
  - Example-
      - H₀: The average marks of students is 70.
  - Alternative Hypothesis (H₁ or Hₐ):
It is the opposite of the null hypothesis and represents the claim that there is an effect or a difference.
  - Example-
     - H₁: The average marks of students is not 70.
  - Difference:
The null hypothesis assumes no change, while the alternative hypothesis suggests a significant change or effect.

3.  Explain the significance level in hypothesis testing and its role in deciding
the outcome of a test.
  - The significance level (α) is the probability of rejecting the null hypothesis when it is actually true. It represents the risk of making a Type I error.
  - Common values of α are 0.05 (5%), 0.01 (1%), and 0.10 (10%).
Role in decision-making:
  - The significance level sets the threshold for deciding whether the test result is statistically significant.
    - If p-value ≤ α → Reject the null hypothesis
    - If p-value > α → Fail to reject the null hypothesis

4.  What are Type I and Type II errors? Give examples of each.
  - A Type I error occurs when the null hypothesis is rejected even though it is actually true. This type of error is also called a false positive.
    - For example, a medical test shows that a person has a disease even though the person is healthy, or concluding that a new medicine is effective when in reality it is not.
  - A Type II error occurs when the null hypothesis is not rejected even though it is actually false. This type of error is known as a false negative.
    - For example, a medical test shows that a person is healthy even though the person actually has the disease, or concluding that a medicine is not effective when it actually works.

5. What is the difference between a Z-test and a T-test? Explain when to use
each.
  - The Z-test and T-test are statistical tests used to test hypotheses about population means.
  - It is used when the sample size is large (n ≥ 30) and the population standard deviation is known. It is based on the normal distribution.
    - Example: A factory knows the population standard deviation of bottle weights and tests whether the average weight of 50 bottles is equal to 1 litre. Since the sample size is large and standard deviation is known, a Z-test is used.
   - A T-test is used when the sample size is small (n < 30) and the population standard deviation is unknown.
   - It is based on the t-distribution and uses degrees of freedom.
     - Example: A teacher wants to check whether the average marks of 15 students are equal to 60, but the population standard deviation is not known. In this case, a T-test is used.

```python
#6. Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.

"""A binomial distribution models the number of successes in a fixed number of independent trials,
where each trial has the same probability of success.
Number of trials (n) = 10 and Probability of success (p) = 0.5"""

import numpy as np
import matplotlib.pyplot as plt

# Generate binomial data
n = 10        # number of trials
p = 0.5       # probability of success
size = 1000  # number of experiments

data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(data, bins=range(0, n+2), edgecolor='black')
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.show()
```

```python
"""8. Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib."""

import numpy as np
import matplotlib.pyplot as plt

# Generate normal data
data = np.random.normal(50, 5, 100)

# Calculate mean
mean = np.mean(data)

# Calculate standard deviation
std = np.std(data)

# Number of observations
n = len(data)

# Z value for 95% confidence interval
z = 1.96

# Confidence interval
lower = mean - z * (std / np.sqrt(n))
upper = mean + z * (std / np.sqrt(n))

print("Mean:", mean)
print("95% Confidence Interval:", lower, "to", upper)

# Histogram
plt.hist(data)
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Normal Distribution")
plt.show()
```

```python
"""9. Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram.
Explain what the Z-scores represent in terms of standard deviations from the mean."""

#A Z-score tells how many standard deviations a value is away from the mean.
#Z = 0 → value is exactly equal to the mean
#Z > 0 → value is above the mean
#Z < 0 → value is below the mean

import numpy as np
import matplotlib.pyplot as plt

# Function to calculate Z-scores
def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data)
    z_scores = (data - mean) / std
    return z_scores

# Sample dataset
data = np.array([45, 50, 52, 48, 47, 51, 49, 53, 46, 50])

# Calculate Z-scores
z_scores = calculate_z_scores(data)

print("Z-scores:", z_scores)

# Plot histogram
plt.hist(z_scores)
plt.xlabel("Z-score values")
plt.ylabel("Frequency")
plt.title("Histogram of Z-scores")
plt.show()
```
